```python

import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(24,18))  
N = 8
ind = np.arange(N) 
width = 0.25
#pos  
xvals = [130,42,10,16,9,64,41,5]
bar1 = plt.bar(ind, xvals, width, color = 'r')

#neg  
yvals = [79,51,74,65,45,103,43,7]
bar2 = plt.bar(ind+width, yvals, width, color='g')

#neutral  
zvals = [15,22,9,12,122,16,13,7]
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')
  
plt.xlabel("\n Catagories",fontsize=23)
plt.ylabel('Numbers',fontsize=23)
plt.title("")
  
plt.xticks(ind+width,['Vaccination','Covid \n Restrictions','Political','Health Effect \n of Covid','News',
                      'Spread of Covid \n in Society','Covid Test/ \n Healthcare','Spam'],fontsize=22)
plt.yticks(fontsize=23)
plt.legend( (bar1, bar2, bar3), ('Positive', 'Negative', 'Neutral'),fontsize=23 )
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Vaccination','Covid \n Restrictions','Political','Health Effect of Covid','News','Spread of Covid \n in Society','Covid Test Healthcare','Spam'
sizes = [224,115,93,93,176,183,97,19]
newList=[x / 1000 for x in sizes]
explode = (0, 0,0,0,0,0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(10,10))
# fig1=plt.figure(figsize=(20,20))
ax1.pie(newList, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90,textprops={'fontsize': 14})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
